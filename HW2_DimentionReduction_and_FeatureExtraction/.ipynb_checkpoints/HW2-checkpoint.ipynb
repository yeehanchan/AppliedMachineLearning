```python
import numpy as np
from matplotlib import pylab as plt
from scipy import misc
import matplotlib.cm as cm
```

```python
train, train_label,test,test_label = [], [], [],[]
#import os 
#os.getcwd()
with open("train.txt") as f:
    for line in f:
        im = misc.imread("../"+line.strip().split()[0])
        train.append(im.reshape(2500,))
        train_label.append(line.strip().split()[1])
train = np.asarray(train,dtype='float')
train_label = np.asarray(train_label,dtype='int')
with open("test.txt") as f:
    for line in f:
        im = misc.imread("../"+line.strip().split()[0])
        test.append(im.reshape(2500,))
        test_label.append(line.strip().split()[1])
test = np.asarray(test,dtype='float')
test_label = np.asarray(test_label,dtype='int')

print(train.shape)
print(test.shape)

plt.imshow(train[10].reshape(50,50),cmap=cm.Greys_r)
plt.show()
        
```

```python
svd = np.linalg.svd(MS_train,True,True)
U = svd[0]
V = svd[2]
S = np.zeros([U.shape[1],V.shape[0]])
j=0
for i in range(0,U.shape[0]):
    S[i][j] = svd[1][i];
    j+=1;
#print(S[539][540])
def r_approximation(r):
    RAE = []
    for i in range(0,r):
#        X_r = np.multiply(np.multiply(U[:,:r],S[:r,:r]),V[:r,:])
        X_r = (U[:,:i].dot(S[:i,:i])).dot(V[:i,:])
#        print(U[:,:i].shape)
#        print("-----------------------------------")
        RAE.append(np.linalg.norm(np.subtract(train,X_r),ord='fro'))
    plt.plot(np.arange(0,r),RAE)
r_approximation(200)
```

```python
def r_feature(r,X):
    V_r = V[:r,]
    return X.dot(np.transpose(V_r))
#    print(np.transpose(V_r).shape)
#F = r_feature(200,train)
#F_test = r_feature(200,test)
```

```python
from sklearn.linear_model import LogisticRegression
def logistic_regression_test(r,train,train_label,test,test_label):
    clf = LogisticRegression()
    accuracy = []
    for i in range(1,r):
        F = r_feature(i,train)
        F_test = r_feature(i,test)
        clf.fit(F,train_label)
#        print(clf.predict(F_test))
        accuracy.append(clf.score(F_test,test_label))
    plt.plot(np.arange(1,r),accuracy)
    return accuracy
a = logistic_regression_test(200,train,train_label,test,test_label)
```
